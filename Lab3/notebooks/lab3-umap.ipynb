```python
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid")
import umap.umap_ as umap
```

```python
digits = load_digits(as_frame=True)
digits_df = digits.frame.sample(n=300, random_state=42)
X_digits = digits_df.drop(columns=['target'])
y_digits = digits_df['target'].astype(str)
X_umap = umap.UMAP(n_components=2, random_state=42).fit_transform(X_digits)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=y_digits, palette="tab10", s=35, ax=ax, legend=False)
ax.set_title("UMAP na digits")
plt.tight_layout()
```
